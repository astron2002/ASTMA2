```python
print(r.content)
```

```python
print(r.url)
```

```python
print(r.status_code)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(r.content,'html.parser')
print(soup.prettify())
```

```python
print(soup.title)
```

```python
print(soup.title.name)
```

```python
print(soup.title.parent.name)
```

```python
s=soup.find('div',class_='text')
content=s.find_all('p')
print(content)
```

```python
for line in content:
    print(line)
    print(line.text)
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
images=[]
for image in soup.select('img'):
    src=image.get('src')
    alt=image.get('alt')
    images.append({'src':src, 'alt':alt})
    
for image in images:
    print(image)
```
