```python
# Function to count word frequencies in a text document
def count_word_frequencies(text):
    # Convert the text to lowercase to ensure case-insensitive counting
    text = text.lower()
    
    # Split the text into words
    words = text.split()
    
    # Dictionary to store word frequencies
    word_freq = {}
    
    # Count the frequency of each word
    for word in words:
        # Remove punctuation characters (you can add more if needed)
        word = word.strip('.,!?()[]{}":;')
        
        # Update the word frequency dictionary
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
    
    return word_freq
```

```python
# Sample text for testing (you can replace this with your own text or read from a file)
sample_text = """
This is a sample text document. It contains several words, and we want to count
the frequency of each word.
The words in this document may repeat, but we are interested in finding the most
common words.
"""
```

```python
# Count word frequencies in the sample text
word_freq = count_word_frequencies(sample_text)
word_freq
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.bar(word_freq.keys(), word_freq.values())
plt.show()
```

```python
sorted_word_freq=sorted(word_freq.items(), key= lambda word:(word[1],word[0]),reverse=True)
sorted_word_freq
```

```python
print(dict(sorted_word_freq))
```

```python
words, counts = zip(*sorted_word_freq)

# Create bar chart
plt.bar(words, counts)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency')
plt.xticks(rotation=90) # Rotate x-axis labels for readability
plt.show()
```
